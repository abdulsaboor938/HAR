```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data without ambient
data = pd.read_csv('reports/bubbly_coconut_2znry3qw.csv', index_col=0)
data.head()
```

```python
# Plot evaluation metrics
plt.figure(figsize=(10, 6))
plt.plot(range(1, 80000, 400), data['precision'], label='Precision')
plt.plot(range(1, 80000, 400), data['recall'], label='Recall')
plt.plot(range(1, 80000, 400), data['f1'], label='F1 Score')
plt.xlabel('Training Examples')
plt.ylabel('Score')
plt.title('Evaluation Metrics Over first 80000 Samples (XGBoost without Ambient Data)')
plt.legend()
plt.grid(True)

# save plot to file
plt.savefig('plots/6.2 xgboost-iter_without.png')
```
